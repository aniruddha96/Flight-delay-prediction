```python
%%time
import datetime, warnings, scipy 
import pandas as pd
import numpy as np
df_all = pd.read_csv('flights.csv', low_memory=False)
```

```python
#df=df_all.head(100000)
df=df_all
print(df.shape)
```

```python
df2=df[['MONTH','DAY','DAY_OF_WEEK','AIRLINE','ORIGIN_AIRPORT','DESTINATION_AIRPORT','AIR_SYSTEM_DELAY','SECURITY_DELAY','AIRLINE_DELAY','LATE_AIRCRAFT_DELAY',
           'WEATHER_DELAY','SCHEDULED_DEPARTURE','SCHEDULED_ARRIVAL']]
```

```python
X=df2[['MONTH','DAY','DAY_OF_WEEK','AIRLINE','ORIGIN_AIRPORT','DESTINATION_AIRPORT','SCHEDULED_DEPARTURE','SCHEDULED_ARRIVAL']]
```

```python
df2['tot_d']=df2['AIR_SYSTEM_DELAY']+df2['SECURITY_DELAY']+df2['AIRLINE_DELAY']+df2['LATE_AIRCRAFT_DELAY']+df2['WEATHER_DELAY']
Y=df2[['tot_d']]
Y.fillna(0,inplace=True)
Y[Y <= 20] = 0
Y[Y > 20] = 1
```

```python
from sklearn import preprocessing
le = preprocessing.LabelEncoder()
le.fit(X['ORIGIN_AIRPORT'])
X['ORIGIN_AIRPORT']=le.transform(X['ORIGIN_AIRPORT'])
le.fit(X['DESTINATION_AIRPORT'])
X['DESTINATION_AIRPORT']=le.transform(X['DESTINATION_AIRPORT'])
le.fit(X['AIRLINE'])
X['AIRLINE']=le.transform(X['AIRLINE'])
```

```python
xtrain=X.head(5000000)
ytrain=Y.head(5000000)

xtest=X.tail(819079)
ytest=Y.tail(819079)
```

```python
from sklearn.ensemble import RandomForestClassifier
model = RandomForestClassifier(verbose=3,n_estimators=20,n_jobs=-1)
model.fit(xtrain,ytrain)
```

```python
from sklearn.metrics import accuracy_score
print(accuracy_score(ytrain, model.predict(xtrain)))
print(accuracy_score(ytest, model.predict(xtest)))
```

```python
from sklearn.metrics import confusion_matrix
confusion_matrix(ytest,model.predict(xtest))
```
